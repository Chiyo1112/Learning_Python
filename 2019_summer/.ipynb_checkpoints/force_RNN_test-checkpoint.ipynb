# 深度学习

## 加载模块包，设置参数

```python
import torch
import torch.nn as nn
import torchvision.datasets as dsets
import torchvision.transforms as transforms
from torch.autograd import Variable
```

```python
#torch.manual_seed(1)#设置随机数种子
learning_rate = 0.01#学习率
num_epochs = 2#训练次数
batch_size = 100#批处理大小
num_classes = 10
layer_num =3
hidden_size = 128
input_size = 28
sequence_length = 28

look_back_number = 3
```

## 创建数据集

### 读入数据并归一化

```python
data_csv = force_all_1.My[10000:20000]
dataset = data_csv.values
dataset = dataset.astype('float32')
max_value = np.max(dataset)
min_value = np.min(dataset)
scalar = max_value - min_value
dataset = list(map(lambda x: x / scalar, dataset))
```

### 创建数据集

想通过前面几个月的流量来预测当月的流量，我们可以将前两个月的流量当做输入，当月的流量当做输出。

同时需要将的数据集分为训练集和测试集，通过测试集的效果来测试模型的性能，这里我们简单的将前面几年的数据作为训练集，后面两年的数据作为测试集。

```python

def create_dataset(dataset, look_back=look_back_number):#改成5个数预测1个数
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
    return np.array(dataX), np.array(dataY)
```

```python
# 创建好输入输出
data_X, data_Y = create_dataset(dataset)
```

## 定义循环神经网络

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(RNN, self).__init()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first = True)
        self.fc = nn.Linear(hidden_size, num_classes)
```

#### 划分训练集和测试集，70% 作为训练集

```python
#train_size = int(len(data_X) * 0.7)
#test_size = len(data_X) - train_size
train_X = data_X[:3500]
train_Y = data_Y[:3500]
test_X = data_X[3500:5000]
test_Y = data_Y[3500:5000]
```

最后，我们需要将数据改变一下形状，因为 RNN 读入的数据维度是 (seq, batch, feature)，所以要重新改变一下数据的维度，这里只有一个序列，所以 batch 是 1，而输入的 feature 就是我们希望依据的几个月份，这里我们定的是两个月份，所以 feature 就是 2.

```python
import torch

train_X = train_X.reshape(-1, 1, look_back_number)
train_Y = train_Y.reshape(-1, 1, 1)
test_X = test_X.reshape(-1, 1, look_back_number)

train_x = torch.from_numpy(train_X)
train_y = torch.from_numpy(train_Y)
test_x = torch.from_numpy(test_X)
test_y = torch.from_numpy(test_Y)

train_x.is_cuda
print(type(train_x))
```

## 定义网络

这里定义好模型，模型的第一部分是一个两层的 RNN，每一步模型接受两个月的输入作为特征，得到一个输出特征。接着通过一个线性层将 RNN 的输出回归到流量的具体数值，这里我们需要用 view 来重新排列，因为 nn.Linear 不接受三维的输入，所以我们先将前两维合并在一起，然后经过线性层之后再将其分开，最后输出结果。

#### 单向RNN

```python
class lstm_reg(nn.Module):
    def __init__(self, input_size, hidden_size, output_size=1, num_layers=layer_num):
        super(lstm_reg, self).__init__()
        
        self.rnn = nn.LSTM(input_size, hidden_size, num_layers) # rnn
        self.reg = nn.Linear(hidden_size, output_size) # 回归
        #self.reg = self.reg.cuda()
        
    def forward(self, x):
        x, _ = self.rnn(x) # (seq, batch, hidden)
        s, b, h = x.shape
        x = x.view(s*b, h) # 转换成线性层的输入格式
        x = self.reg(x)
        x = x.view(s, b, -1)
        return x
```

#### 双向RNN

### 判断是否cuda

```python
if torch.cuda.is_available():
    net = lstm_reg(look_back_number, 4).cuda()#网络转cuda()
    train_x = train_x.cuda()
    train_y = train_y.cuda()
    test_x = test_x.cuda()
    test_y = test_y.cuda()
else:
    net = lstm_reg(look_back_number, 4)

var_x = Variable(train_x)#数据转化为Variable给PyTorch
var_y = Variable(train_y)#数据转化为Variable给PyTorch
```

### 损失函数

```python
criterion = nn.MSELoss()#标准
#criterion = nn.CrossEntropyLoss()
```

### 优化器

```python
optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate) 
```

## 读取网络

## 开始训练

```python
import time
start = time.time()
calculate_step = 10000
loss_x, loss_y = [], []
# 开始训练
for e in range(calculate_step):

    # 前向传播
    out = net(var_x)#前向算法计算输出
    loss = criterion(out, var_y)#计算输出与真实输出的loss
    
    # 反向传播
    optimizer.zero_grad()#梯度清零
    loss.backward()#后向算法，计算梯度
    optimizer.step()#应用梯度、更新参数
    
    loss_x.append(e)
    loss_y.append(loss.data)#收集每步的loss进行绘图

    if (e + 1) % 1000 == 0: # 每 100 次输出结果
        print('Epoch: {}, Loss: {:.5f}'.format(e + 1, loss.data))
end = time.time() - start
print('迭代{}步用时{}秒'.format(calculate_step, end))
```
