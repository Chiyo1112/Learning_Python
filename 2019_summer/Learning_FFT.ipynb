# 一类

```python
from scipy import fftpack
def DoFFT(data, freq):
    n = len(data)
    yf = abs(np.fft.fft(data)) / n
    xf = np.fft.fftfreq(n, d=1/freq)
    return xf, yf
```

```python
import numpy as np
x = np.linspace(0,7,700)
y = np.sin(2 * np.pi * x)
```

```python
import matplotlib.pylab as plt
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
freq = 100
z = DoFFT(y, freq)
```

```python
plt.plot(z[0], z[1])
plt.xlim(-2,2)
```

# 二类

```python
def DoFFT(data, freq):
    n = len(data)
    yf = 2* abs(np.fft.fft(data)) / n
    xf = fftpack.rfftfreq(n, d=1/freq)*2
    return xf, yf
```

```python
import numpy as np
x = np.linspace(0,7,700000)
y = np.sin(2 * np.pi * x)
```

```python
freq = 100000
z = DoFFT(y, freq)
plt.plot( z[1])
plt.xlim(0,20)
```
