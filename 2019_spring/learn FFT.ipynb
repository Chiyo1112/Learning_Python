N点FFT能精确计算的频率

假设取样频率为fs, 取波形中的N个数据进行FFT变换。那么这N点数据包含整数个周期的波形时，FFT所计算的结果是精确的。于是能精确计算的波形的周期是: n*fs/N。对于8kHz取样，512点FFT来说，8000/512.0 = 15.625Hz，前面的156.25Hz和234.375Hz正好是其10倍和15倍。

```python
import numpy as np
import pylab as pl

sampling_rate = 8000#数字信号的取样频率
fft_size = 512      #FFT的长度

#产生1秒钟的取样时间，t中每个值直接表示取样点的时间，间隔为取样周期1/sampline_rate
t = np.arange(0, 1.0, 1.0/sampling_rate)

#用取样时间数组t调用函数计算出波形数据，两个正弦波的叠加，一个频率是156.25Hz，一个是234.375Hz：
x = np.sin(2*np.pi*156.25*t)  + 2*np.sin(2*np.pi*234.375*t)

#下面从波形数据x中截取fft_size个点进行fft计算。np.fft库中提供了一个rfft函数，它方便我们对实数信号进行FFT计算。
#根据FFT计算公式，为了正确显示波形能量，还需要将rfft函数的结果除以fft_size：
xs = x[:fft_size]
xf = np.fft.rfft(xs)/fft_size

#rfft函数的返回值是N/2+1个复数，分别表示从0(Hz)到sampling_rate/2(Hz)的N/2+1点频率的成分。
#可以通过np.linspace计算出返回值中每个下标对应的真正的频率：
freqs = np.linspace(0, sampling_rate/2, fft_size/2+1)

#最后我们计算每个频率分量的幅值，并通过 20*np.log10() 将其转换为以db单位的值。
#为了防止0幅值的成分造成log10无法计算，我们调用np.clip对xf的幅值进行上下限处理：
xfp = 20*np.log10(np.clip(np.abs(xf), 1e-20, 1e100))

#将时域波形和频域波形绘制出来
pl.figure(figsize=(8,4))
pl.subplot(211)
pl.plot(t[:fft_size], xs)
pl.xlabel(u"时间(秒)")
pl.title(u"156.25Hz和234.375Hz的波形和频谱")
pl.subplot(212)
pl.plot(freqs, xfp)
pl.xlabel(u"频率(Hz)")
pl.subplots_adjust(hspace=0.4)
pl.show()
```

## FFT中有用的数值

```python
import numpy as np
import pylab as pl
x = np.random.rand(8)
```

```python
xf = np.fft.fft(x)
print(xf)
plt.plot(np.real(xf), color='blue')
plt.plot(np.imag(xf), color='red')
plt.plot(np.abs(xf), color = 'black')

plt.grid()
```

```python
z = np.fft.ifft(xf)
print(z)
plt.plot(np.real(z), color='blue')
plt.plot(np.imag(z), color='red')
plt.plot(np.abs(z), color = 'black')
plt.plot(x,lw=10, color = 'yellow',alpha = 0.5)
plt.grid()
```

```python
x = np.arange(0, 2*np.pi, 2*np.pi/8)
y = np.sin(x)
yf = np.fft.fft(y)/len(y)
plt.plot(np.real(yf), color='blue',label='real')
plt.plot(np.imag(yf), lw=5, alpha=0.5,color='red', label='imag')
plt.plot(np.abs(yf), color = 'black',label='abs')
plt.plot(yf,lw=20, color = 'green',label='fft',alpha = 0.2)
plt.plot(y, color = 'pink',label='y')
plt.legend()
plt.grid()
```

```python
x = np.arange(0, 2*np.pi, 2*np.pi/8)
y = np.cos(x)
yf = np.fft.fft(y)/len(y)
plt.plot(np.real(yf), color='blue',label='real')
plt.plot(np.imag(yf), color='red', label='imag')
plt.plot(np.abs(yf), color = 'black',label='abs')
plt.plot(y, color = 'pink',label='y')
plt.legend()
plt.grid()
```
