```python
class TDMS_Vibrition_Merge():
    def __init__(self, file_address, file_name, group_name = 'Measured Data'):
        self.file_address = file_address
        self.file_name = file_name

        n = 0
        for i in file_name:
            real_path = r"{}\{}".format(file_address, file_name[n])
            TDMS_read = TdmsFile(r'{}'.format(real_path))

            channel_x_0 = TDMS_read.object(group_name, 'x_0')
            channel_y_0 = TDMS_read.object(group_name, 'y_0')
            channel_z_0 = TDMS_read.object(group_name, 'z_0')
            channel_x_1 = TDMS_read.object(group_name, 'x_1')
            channel_y_1 = TDMS_read.object(group_name, 'y_1')
            channel_z_1 = TDMS_read.object(group_name, 'z_1')

            getdata_x_0 = channel_x_0.data
            getdata_y_0 = channel_y_0.data
            getdata_z_0 = channel_z_0.data
            getdata_x_1 = channel_x_1.data
            getdata_y_1 = channel_y_1.data
            getdata_z_1 = channel_z_1.data

            if n == 0:
                self.data_x_0 = getdata_x_0
                self.data_y_0 = getdata_y_0
                self.data_z_0 = getdata_z_0
                self.data_x_1 = getdata_x_1
                self.data_y_1 = getdata_y_1
                self.data_z_1 = getdata_z_1

            else:
                self.data_x_0 = np.concatenate((self.data_x_0, getdata_x_0))
                self.data_y_0 = np.concatenate((self.data_y_0, getdata_y_0))
                self.data_z_0 = np.concatenate((self.data_z_0, getdata_z_0))
                self.data_x_1 = np.concatenate((self.data_x_1, getdata_x_1))
                self.data_y_1 = np.concatenate((self.data_y_1, getdata_y_1))
                self.data_z_1 = np.concatenate((self.data_z_1, getdata_z_1))

            n += 1
            
    def x_0(self):
        return np.array(self.data_x_0)
    
    def y_0(self):
        return np.array(self.data_y_0)
    
    def z_0(self):
        return np.array(self.data_z_0)
    
    def x_1(self):
        return np.array(self.data_x_1)
    
    def y_1(self):
        return np.array(self.data_y_1)

    def z_1(self):
        return np.array(self.data_z_1)
```

```python
class TDMS_Power_Merge():
    def __init__(self, file_address, file_name, group_name = 'Measured Data'):
        self.file_address = file_address
        self.file_name = file_name

        n = 0
        for i in file_name:
            real_path = r"{}\{}".format(file_address, file_name[n])
            TDMS_read = TdmsFile(r'{}'.format(real_path))

            channel_power = TDMS_read.object(group_name, 'Power')


            getdata_power = channel_power.data

            if n == 0:
                self.data_power = getdata_power
            else:
                self.data_power = np.concatenate((self.data_power, getdata_power))

            n += 1
            
    def power(self):
        return np.array(self.data_power)
```

```python
class Sample_it():
    def __init__(self, data, number):
        
        len_0 = len(data)
        len_1 = len_0 // number
        len_2 = np.arange(len_1)
        len_3 = len_2 * number
        
        self.data_new = data.take(len_3)
    def sampled(self):
        return self.data_new
```

# FFT

```python
from scipy import fftpack
def DoFFT(data):
    n = len(data)
    yf = 2 * abs(np.fft.fft(data)) / n
    xf = fftpack.rfftfreq(n, d=1/1000)*2
    return xf, yf
```

```python
m =    500000
n =    600000
p = 115400000
q = 115500000
```

```python
class Show_Them():
    def __init__(self, data_new, data_old, start_new, start_old, step):
       
        xf_new = DoFFT(data_new[start_new : start_new + step])[0]
        yf_new = DoFFT(data_new[start_new : start_new + step])[1]
        xf_old = DoFFT(data_old[start_new : start_new + step])[0]
        yf_old = DoFFT(data_old[start_old : start_old + step])[1]
        n =step / 2
        
        plt.figure(figsize=(16,10))
        plt.subplot(2,1,1)
        plt.plot(data_new[start_new : start_new + step], label = 'new tool',color = 'deepskyblue', alpha = 0.8)
        plt.plot(data_old[start_old : start_old + step], label = 'broken tool',color = 'red', alpha = 0.8)
        plt.title('The vibration changes by x_0', fontsize = 14)
        plt.legend(loc='best', framealpha = 0.9)
        plt.grid()
        plt.xlabel('Samples', fontsize=10)
        plt.ylabel('Amp (g)',  fontsize=10)
        plt.tick_params(axis = 'both', labelsize = 10)
        plt.xlim(0, 2*n)
        
        plt.subplot(2,1,2)
        plt.plot(xf_new, yf_new, label = 'new tool',color = 'deepskyblue', alpha = 0.8)
        plt.plot(xf_old, yf_old, label = 'broken tool',color = 'red', alpha = 0.8)
        plt.title('The vibration changes by x_0', fontsize = 14)
        plt.legend(loc='best', framealpha = 0.9)
        plt.grid()
        plt.xlabel('frequency (un)', fontsize=10)
        plt.ylabel('FFT',  fontsize=10)
        plt.xlim(0, 40)
        plt.tick_params(axis = 'both', labelsize = 10)
        
Show_Them(data_new = data_vibration_20190513_0_x_0[m:n],
          data_old = data_vibration_20190513_0_x_0[p:q],
          start_new = 0, 
          start_old = 0,
          step = m-n)
```
